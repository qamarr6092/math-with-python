### Dot Product of two vectors

Let $\mathbf{v}$ and $\mathbf{w}$ be two column vectors:

$$\mathbf{v}^T \mathbf{w} = \begin{bmatrix} v_1 & v_2 & \dots & v_n \end{bmatrix} \begin{bmatrix} w_1 \\ w_2 \\ \vdots \\ w_n \end{bmatrix} = v_1w_1 + v_2w_2 + \dots + v_nw_n$$

Or in short summation notation:

$$\mathbf{v}^T \mathbf{w} = \sum_{i=1}^{n} v_i w_i$$

### Projection

This formula calculates the single number ($\beta$) that tells you how much to stretch or shrink the line vector $\mathbf{a}$ to match the length of the shadow.

Using matrix transpose notation:
$$\beta = \frac{\mathbf{a}^T\mathbf{b}}{\mathbf{a}^T\mathbf{a}}$$

Using standard dot product notation:
$$\beta = \frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{a} \cdot \mathbf{a}}$$

---

### 2. The Projection Vector Formula ($\mathbf{p}$)

To get the actual shadow arrow ($\mathbf{p}$), you multiply the scalar $\beta$ by your ground vector $\mathbf{a}$:
$$\mathbf{p} = \beta\mathbf{a}$$

Substituting the full formula for $\beta$ gives the complete orthogonal projection formula:
$$\mathbf{p} = \left( \frac{\mathbf{a}^T\mathbf{b}}{\mathbf{a}^T\mathbf{a}} \right) \mathbf{a}$$

---

### 3. The Projection Matrix Formula ($P$)

If you want to build a standalone flattening matrix $P$ that separates vector $\mathbf{b}$ entirely:
$$P = \frac{\mathbf{a}\mathbf{a}^T}{\mathbf{a}^T\mathbf{a}}$$

Once you compute matrix $P$, you find the projection vector by doing a standard matrix multiplication:
$$\mathbf{p} = P\mathbf{b}$$

## Linear Combination

A **linear combination** (or **weighted linear combination**) is the operation of scaling a set of vectors by coefficients (weights) and adding them together:

$$\mathbf{y} = c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \dots + c_n\mathbf{v}_n$$

* **Example:** The target vector $\mathbf{t} = \begin{bmatrix} 4 \\ 5 \end{bmatrix}$ is a linear combination of the standard unit vectors $\mathbf{i} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ and $\mathbf{j} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$ with weights $c_1 = 4$ and $c_2 = 5$:
$$\mathbf{t} = 4\mathbf{i} + 5\mathbf{j}$$

---

The Homogeneous Equation
A **homogeneous equation** is any vector equation where a linear combination is set equal to the zero vector $\mathbf{0}$ (the origin):

$$x_1\mathbf{v}_1 + x_2\mathbf{v}_2 + \dots + x_n\mathbf{v}_n = \mathbf{0}$$

Derivation from Geometric Dependence:
If a third vector $\mathbf{v}_3$ lies flat in the 2D plane spanned by $\mathbf{v}_1$ and $\mathbf{v}_2$, it can be written as a combination of them:
$$\mathbf{v}_3 = 2\mathbf{v}_1 + 3\mathbf{v}_2$$

Subtracting $\mathbf{v}_3$ from both sides balances the right side to zero, naturally yielding the homogeneous form:
$$2\mathbf{v}_1 + 3\mathbf{v}_2 - 1\mathbf{v}_3 = \mathbf{0}$$

---

Trivial vs. Non-Trivial Solutions
Solving a homogeneous equation means finding the specific weights $x_i$ that satisfy it:

* **Trivial Solution:** Setting all weights to zero ($x_1 = 0, x_2 = 0, \dots, x_n = 0$). This is the obvious solution that works for any set of vectors because you don't move at all.
* **Non-Trivial Solution:** A solution where at least one weight is non-zero (e.g., $x_1=2, x_2=3, x_3=-1$). This proves you can travel along the vectors and form a complete closed loop that returns to the origin.

---

Linear Independence vs. Dependence
The availability of non-trivial solutions defines the structural relationship between the vectors:

* **Linearly Independent:** The *only* way to satisfy the homogeneous equation is the trivial solution. Every vector points in a brand-new direction (e.g., three vectors where the third sticks straight up out of the plane like a flagpole). The trivial solution is **unique**.
* **Linearly Dependent:** At least one non-trivial solution exists. There is spatial redundancy; at least one vector lies completely within the span of the others (e.g., a third vector lying flat on a 2D floor plane). The trivial solution is **not unique** because infinitely many non-zero combinations also equal zero.

---

 Rank of a Matrix
The **rank** of a matrix is the number of **truly unique, linearly independent directions** contained within its columns (or rows). It represents the true geometric dimensionality of the vector space it spans.

* **Rank 1:** All columns collapse onto a 1D line.
* **Rank 2:** Columns spread out to form a flat 2D plane.
* **Rank 3:** Columns completely fill out a 3D volume.

Key Properties:
* **Full Rank:** Every single row and column is linearly independent.
* **Rank-Deficient:** Redundancy exists; at least one column can be built from the others.
* **The Invariant:** $\text{Column Rank} = \text{Row Rank}$ always.

---

Finding Rank via Row Echelon Form
When performing Gaussian elimination to reduce a matrix to Row Echelon Form:
$$\text{Rank} = \text{Number of non-zero rows (rows containing pivots)}$$

If a matrix is rank-deficient, a row on the left side of an augmented matrix will completely flatten into zeros: `[0 0 ... 0]`. This leads to two critical bottom-row scenarios in a system of linear equations:

Scenario A: The $0 = 0$ Result (Infinite Solutions)
$$0x_1 + 0x_2 + \dots + 0x_n = 0 \implies 0 = 0$$
* **Meaning:** An entire row was a perfect, valid duplicate of other instructions. The system is **consistent** but redundant, creating free variables and an infinite number of overlapping intersection points.

Scenario B: The $0 = \text{Non-Zero}$ Result (No Solution)
$$0x_1 + 0x_2 + \dots + 0x_n = 4 \implies 0 = 4$$
* **Meaning:** The equations point in a redundant direction but physically contradict each other (e.g., parallel lines or planes that are separated). The system is **inconsistent** and has zero solutions.

### Basis of vector space

 Span (The Destination)
The infinite collection of all possible linear combinations of a set of vectors.
$$\text{Span}(\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k) = \{ c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \dots + c_k\mathbf{v}_k \}$$

Spanning (The Action)
The act of completely covering a space. A set of vectors spans a space if they can reach every single point inside it, even if some vectors are redundant.

Subspace (The Playground)
A flat region living inside a larger space that passes exactly through the origin $\mathbf{0}$ and obeys two strict rules:
* **Closed under addition:** $\mathbf{u} + \mathbf{w} \in H$
* **Closed under scalar multiplication:** $c\mathbf{u} \in H$

Basis (The Minimalist Toolkit)
A perfect coordinate system for a subspace. It is a "Goldilocks" set of vectors that must satisfy two criteria:
1. They **Span** the subspace (enough to reach everything).
2. They are **Linearly Independent** (absolutely zero redundancy).

* *Example:* A 2D plane requires a basis of exactly **2** independent vectors. 3 vectors is overkill (not a basis); 1 vector is not enough (not a basis).

 Dimension
The locked number of vectors inside a subspace's basis.
$$\text{dim}(H) = \text{Number of vectors in its basis} = \text{Matrix Rank}$$

### K means clustering

````python
r = np.random.uniform( size = (1000,2) ) 
k =4
i = np.random.choice(len(r), size = k, replace = False ) 
centroids = r[i]
plt.figure(figsize =(8,8))
plt.scatter(r[:,0], r[:,1])
plt.scatter(centroids[:,0], centroids[:,1] ,marker='x',alpha=0.9,color='red',s=400,linewidths =3)
plt.show()
````

````python
difference = r[:, np.newaxis, :] - centroids
squared_difference = difference ** 2
summed_squared_distance = np.sum(squared_difference, axis=2)
distances = np.sqrt(summed_squared_distance)
labels = np.argmin(distances, axis=1)
plt.figure(figsize=(8,8))
plt.scatter(r[:,0], r[:,1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:,0], centroids[:,1], marker='x', alpha=0.9, color='red', s=400, linewidths=3)
plt.show()
````

````python
num_iterations = 10
for iteration in range(num_iterations):
    difference = r[:, np.newaxis, :] - centroids
    squared_difference = difference ** 2
    summed_squared_distance = np.sum(squared_difference, axis=2)
    distances = np.sqrt(summed_squared_distance)
    
    labels = np.argmin(distances, axis=1)
    
    for idx in range(k):
        cluster_points = r[labels == idx]
 
        if len(cluster_points) > 0:
            centroids[idx] = np.mean(cluster_points, axis=0)

plt.figure(figsize=(8,8))
plt.scatter(r[:,0], r[:,1], c=labels, cmap='viridis', s=30, alpha=0.6)
plt.scatter(centroids[:,0], centroids[:,1], marker='x', alpha=0.9, color='red', s=400, linewidths=4)
plt.title("After 10 Iterations")
plt.show()
````

### Matrix Multiplication

A matrix transformation does not just move a single vector; it transforms the **entire coordinate space linearly**.

When a matrix transforms space, the entire coordinate grid warps, stretches, rotates, or shears. However, because the transformation is **linear**, two rules must be preserved:
1. The origin $(0,0)$ must remain perfectly fixed in place.
2. All grid lines must remain straight and evenly spaced.

Because the grid lines stay straight and parallel, we don't need to track every single point in the universe. We only need to know where the standard basis vectors—$\hat{i}$ (unit vector along the x-axis) and $\hat{j}$ (unit vector along the y-axis)—land after the transformation. 

The columns of a transformation matrix are nothing more than the **final landing coordinates** of these basis vectors.

$$A = \begin{bmatrix} \mid & \mid \\ T(\hat{i}) & T(\hat{j}) \\ \mid & \mid \end{bmatrix} = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$$

***

Before any transformation occurs, a vector is defined by its coordinates relative to the standard, un-warped background grid. For example, the vector $\mathbf{v} = \begin{bmatrix} 2 \\ 3 \end{bmatrix}$ is a linear combination stating: *"Take 2 steps in the direction of $\hat{i}$ and 3 steps in the direction of $\hat{j}$."*

$$\mathbf{v} = 2\hat{i} + 3\hat{j} = 2\begin{bmatrix} 1 \\ 0 \end{bmatrix} + 3\begin{bmatrix} 0 \\ 1 \end{bmatrix} = \begin{bmatrix} 2 \\ 3 \end{bmatrix}$$

***

When the transformation matrix moves the entire grid, your vector $\mathbf{v}$ rides along with it. 

If you look at the "shadow" of the original grid, the vector has physically moved to a brand-new spot in space. But **internally, its relationship to the warped grid lines has not changed at all.** It is still the exact same linear combination! 

Instead of taking 2 steps along the *original* $\hat{i}$ and 3 steps along the *original* $\hat{j}$, it now takes **2 steps along the *transformed* $\hat{i}$ and 3 steps along the *transformed* $\hat{j}$**.

$$T(\mathbf{v}) = 2 \cdot T(\hat{i}) + 3 \cdot T(\hat{j})$$

Using our matrix columns, the final transformation calculation unfolds like this:

$$\begin{bmatrix} a & b \\ c & d \end{bmatrix} \begin{bmatrix} 2 \\ 3 \end{bmatrix} = 2\begin{bmatrix} a \\ c \end{bmatrix} + 3\begin{bmatrix} b \\ d \end{bmatrix} = \begin{bmatrix} 2a + 3b \\ 2c + 3d \end{bmatrix}$$

Matrix multiplication is not a robotic rule of "rows times columns." It is a geometric recipe: **The vector's original coordinates act as scaling weights, telling you how to combine the transformed basis vectors to locate your final destination.**

````python
# Creating a scalar matrix, that transforms scaled vectors
I = np.eye(2)
A = 2*I
v = np.array([3,4]) 
v2 = np.dot(A,v)
print(f'Before transformation it is = \n\n{ v[0] }x{I[:,0]}  + {v[1]}x{I[:,1]}\n') 

print(f'After transformation it is = \n\n { v[0] }x{A[:,0]}  + {v[1]}x{A[:,1]}')
````

````python
plt.figure(figsize=(8,8)) 
plt.quiver(0,0 , v[0], v[1] ,angles = 'xy',scale_units='xy' , scale= 1, color ='blue' ,label='Vector')
plt.quiver(0,0 , v2[0], v2[1] ,angles = 'xy',scale_units='xy' , scale= 1, color ='red' ,label='After Transformation')
plt.xlim([-2,10])
plt.ylim([-2,10])
plt.axhline(0, color='black', linewidth=1.5) 
plt.axvline(0, color='black', linewidth=1.5) 
plt.legend()
plt.grid( True)
plt.show()
````

### Matrix addition and subtraction

Unlike matrix multiplication—which represents a sequential chain reaction (one transformation happening *after* another)—matrix addition and subtraction describe events happening **simultaneously, at the exact same instant.**

---
When you add two matrices, you are blending two geometric transformations together. Applied to a vector $\mathbf{v}$, the space distributes perfectly:

$$(A + B)\mathbf{v} = A\mathbf{v} + B\mathbf{v}$$


Imagine the coordinate grid is a stretchy sheet of rubber. 
* Matrix $A$ is a force pulling space to the right.
* Matrix $B$ is a force pulling space upward.

When you compute $A + B$, you turn both forces on **at the exact same instant**. The grid lines don't shift right and *then* shift up; instead, every point in space takes a direct, compromise path to its final destination based on the combined pull.

Mathematically, you are performing standard vector addition on the landing spots of the basis vectors $\hat{i}$ and $\hat{j}$:

$$\begin{bmatrix} a & b \\ c & d \end{bmatrix} + \begin{bmatrix} e & f \\ g & h \end{bmatrix} = \begin{bmatrix} \mathbf{a+e} & \mathbf{b+f} \\ \mathbf{c+g} & \mathbf{d+h} \end{bmatrix}$$

* The new landing spot for $\hat{i}$ is the vector sum of where $A$ sent it and where $B$ sent it: $\begin{bmatrix} a \\ c \end{bmatrix} + \begin{bmatrix} e \\ g \end{bmatrix}$.
* The new landing spot for $\hat{j}$ is the vector sum of its two destinations: $\begin{bmatrix} b \\ d \end{bmatrix} + \begin{bmatrix} f \\ h \end{bmatrix}$.

---
Matrix subtraction also happens instantly. Applied to a vector $\mathbf{v}$, it looks like this:

$$(A - B)\mathbf{v} = A\mathbf{v} - B\mathbf{v}$$

Subtraction acts as a **comparative filter**. It strips away what two transformations have in common to isolate the exact spatial difference between them. 

Imagine you are tracking a physical system:
* Matrix $B$ is the **ideal, perfect transformation** you expected (e.g., a smooth $45^\circ$ rotation).
* Matrix $A$ is what your **imperfect, real-world hardware actually did** (e.g., a shaky, slightly warped rotation).

By calculating the subtraction grid **$A - B$**, you completely subtract away the intended rotation. What is left over? The matrix $A - B$ maps **pure distortion**. It shows you *only* the error vectors showing how far off the real-world grid lines drifted from the ideal grid lines.

#### What happens to the basis vectors?
Instead of adding the landing spots, you are finding the vector pointing **from $B$'s landing spot directly to $A$'s landing spot**:

$$\begin{bmatrix} a & b \\ c & d \end{bmatrix} - \begin{bmatrix} e & f \\ g & h \end{bmatrix} = \begin{bmatrix} \mathbf{a-e} & \mathbf{b-f} \\ \mathbf{c-g} & \mathbf{d-h} \end{bmatrix}$$

* The first column ($\hat{i}$) becomes a vector pointing from where $B$ wanted it to go, straight to where $A$ actually put it.
* The second column ($\hat{j}$) becomes a vector pointing from $B$'s target spot to $A$'s actual spot.

---


* **Multiplication ($AB$)** $\rightarrow$ A timeline of events. Do $B$, **then** do $A$.
* **Addition ($A+B$)** $\rightarrow$ A blend of forces. Apply $A$ and $B$ **simultaneously** to find a new compromise space.
* **Subtraction ($A-B$)** $\rightarrow$ A comparison of spaces. Measure the **instantaneous distance** between where $B$ lands space and where $A$ lands space.

## Determinant 

The determinant of a matrix $A$ (written as $\det(A)$ or $|A|$) is a single scalar value that acts as the **scaling factor of space** under a linear transformation. 

---
Before any transformation occurs, the standard basis vectors $\hat{i}$ (one unit along the x-axis) and $\hat{j}$ (one unit along the y-axis) form a perfect $1 \times 1$ **unit square** with an area of exactly $1$.

When a matrix transformation $A$ is applied to space, that perfect unit square gets knocked over and warped into a slanted **parallelogram**. The area of this new transformed parallelogram is exactly equal to the magnitude of the determinant.

$$\text{Area of Transformed Shape} = |\det(A)| \times \text{Original Area}$$

* If $\det(A) = 6$, every single shape in that 2D space has its area scaled up by a factor of 6.
* If $\det(A) = 1$, the grid lines may stretch or rotate, but the absolute area of any shape remains perfectly preserved.
* If $\det(A) = 0.5$, the transformation compresses space, cutting all areas exactly in half.

---
The geometric definition scales perfectly as you add more axes:
* **In 2D Space:** $\det(A)$ measures the change in a 2D **area**.
* **In 3D Space:** Instead of a unit square, you begin with a 3D **unit cube** formed by $\hat{i}$, $\hat{j}$, and $\hat{k}$ with a volume of exactly $1$. 

After the transformation, the cube warps into a tilted 3D box called a **parallelepiped**. The volume of this new box is exactly equal to the magnitude of the $3 \times 3$ determinant.

$$\text{Volume of Transformed Box} = |\det(A)|$$

---

If you calculate a determinant and get a negative number (e.g., $\det(A) = -2$), it does not mean the physical area or volume is negative. 

The negative sign tells you that the transformation **flipped the orientation of space**—effectively turning the coordinate system inside out like a mirror image or a glove turned inside out.

* **Positive Orientation ($\det(A) > 0$):** If you curl the fingers of your right hand from $\hat{i}$ to $\hat{j}$, your thumb points up along the positive z-axis.
* **Negative Orientation ($\det(A) < 0$):** After the transformation, the vector $\hat{j}$ has landed on the opposite side of $\hat{i}$. Curling your fingers from the new $\hat{i}$ to the new $\hat{j}$ now forces your thumb to point downwards. Space has been flipped.

---
When the determinant of a matrix is exactly zero, it means the transformation completely **squishes, flattens, and collapses space**, destroying at least one entire dimension.

* **In 2D:** A 2D plane gets flattened down into a 1D line or crushed into a 0D single point. The area of the unit square drops to exactly $0$.
* **In 3D:** An entire 3D room gets flattened down into a flat 2D plane, a 1D line, or a 0D point. The volume of the unit cube drops to exactly $0$.

This flattening behavior represents a critical milestone in linear algebra because it connects geometry directly to numerical computation:

1. **Information Loss:** Once an entire 3D space is squished flat onto a 2D plane, data is permanently lost. You cannot uniquely "un-squish" a flat 2D shadow back into its original 3D structure because infinitely many points collapsed into the exact same spots.
2. **No Matrix Inverse ($A^{-1}$):** Because the transformation cannot be reversed uniquely, a matrix with a determinant of zero **cannot be inverted**. This is why a matrix where $\det(A) = 0$ is explicitly called a **singular** or non-invertible matrix.
3. **Impact on Systems of Equations ($A\mathbf{x} = \mathbf{b}$):** * If the transformation $A$ flattens the world onto a 2D plane, and your target vector $\mathbf{b}$ lives outside of that flat landing pad, the equation is impossible to satisfy $\rightarrow$ **Zero Solutions**.
   * If your target vector $\mathbf{b}$ happens to lie perfectly inside that flat landing pad, then an entire infinite line of points will collapse directly onto it $\rightarrow$ **Infinitely Many Solutions**.

A determinant of 4 only guarantees one thing: If you draw any shape (a circle, a square, a triangle) in your original space, after the transformation, the area of that shape will be exactly 4 times larger not the vector

Let the transformation matrix be defined as:
$$A = \begin{bmatrix} 3 & 6 \\ 2 & 4 \end{bmatrix}$$

Let us take a general, arbitrary 2D input vector:
$$\mathbf{v} = \begin{bmatrix} x \\ y \end{bmatrix}$$

When we perform the matrix multiplication to find the transformed vector, we apply the standard row-column combination:
$$A\mathbf{v} = \begin{bmatrix} 3 & 6 \\ 2 & 4 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 3x + 6y \\ 2x + 4y \end{bmatrix}$$

This calculation gives us the explicit coordinates of our new transformed vector:
$$X_{\text{new}} = 3x + 6y$$
$$Y_{\text{new}} = 2x + 4y$$

The standard formula for the slope of a vector on a 2D Cartesian plane is the ratio of its vertical component to its horizontal component:
$$\text{Slope} = \frac{Y_{\text{new}}}{X_{\text{new}}}$$

Substituting our derived coordinate values into this slope equation yields:
$$\text{Slope} = \frac{2x + 4y}{3x + 6y}$$

To simplify this fraction, we can factor out a $2$ from the numerator and a $3$ from the denominator:
$$\text{Slope} = \frac{2(x + 2y)}{3(x + 2y)}$$

Because the algebraic expression $(x + 2y)$ appears perfectly in both the numerator and the denominator, they completely cancel each other out for any vector change:
$$\text{Slope} = \frac{2}{3}$$

### Conclusion
This algebraic cancellation demonstrates that no matter what values of $x$ and $y$ you pass into the matrix, the final coordinate ratio is completely locked. The entire 2D grid is structurally crushed onto a single continuous line with a constant slope of:
$$\text{Slope} \approx 0.6667$$

## Matrix space

### Column space

The **Column Space** of a matrix $A$ (written as $\text{Col } A$) is the set of **all possible output vectors** you can ever get by multiplying any input vector $\mathbf{x}$ by the matrix $A$. 

Algebraically, if your matrix has columns $\mathbf{v}_1$ and $\mathbf{v}_2$:
$$A = \begin{bmatrix} | & | \\ \mathbf{v}_1 & \mathbf{v}_2 \\ | & | \end{bmatrix}$$

Multiplying it by an input vector $\mathbf{x} = \begin{bmatrix} c_1 \\ c_2 \end{bmatrix}$ is the exact same as taking a **linear combination** of those columns using the input values as weights:
$$A\mathbf{x} = c_1\mathbf{v}_1 + c_2\mathbf{v}_2$$

* **In the Column Space:** If a target vector $\mathbf{b}$ can be reached by choosing the right weights ($c_1, c_2$), then $\mathbf{b}$ lives inside the Column Space. The system $A\mathbf{x} = \mathbf{b}$ has a solution.
* **Outside the Column Space:** If a target vector $\mathbf{b}$ is physically impossible to reach no matter what weights you pick, it lives outside the Column Space. The system $A\mathbf{x} = \mathbf{b}$ has no solution.

---
Let's analyze the matrix we used in our discussion:
$$A = \begin{bmatrix} 3 & 6 \\ 2 & 4 \end{bmatrix}$$

Notice that Column 2 is just a scalar multiple of Column 1:
$$\begin{bmatrix} 6 \\ 4 \end{bmatrix} = 2 \times \begin{bmatrix} 3 \\ 2 \end{bmatrix}$$

Because these two vectors point in the exact same direction, they are *linearly dependent*. They are trapped on the same infinite line running through the origin. 

Let's trace what happens when we feed an input vector $\mathbf{x} = \begin{bmatrix} -2 \\ 3 \end{bmatrix}$ into this matrix.

1. **The Input Vector:** $\begin{bmatrix} -2 \\ 3 \end{bmatrix}$ lives in the original, full 2D space. It does *not* have to be on our collapsed line.
2. **The Transformation Action:** We multiply the matrix by the input:
   $$-2\begin{bmatrix} 3 \\ 2 \end{bmatrix} + 3\begin{bmatrix} 6 \\ 4 \end{bmatrix} = \begin{bmatrix} -6 \\ -4 \end{bmatrix} + \begin{bmatrix} 18 \\ 12 \end{bmatrix} = \begin{bmatrix} 12 \\ 8 \end{bmatrix}$$
3. **The Output Result:** Look at the resulting output vector $\begin{bmatrix} 12 \\ 8 \end{bmatrix}$. Notice that $12 \times \frac{2}{3} = 8$. This output vector is perfectly trapped on the 1D line defined by the column $\begin{bmatrix} 3 \\ 2 \end{bmatrix}$.

No matter what 2D vector you feed into this matrix, the output will **always** land somewhere on the line $y = \frac{2}{3}x$. Therefore, the **Column Space of this matrix is a 1D line**, not the whole 2D plane.

What if we want to find an input that outputs the vector $\mathbf{b} = \begin{bmatrix} 4 \\ -9 \end{bmatrix}$?
We set up the linear combination:
$$c_1\begin{bmatrix} 3 \\ 2 \end{bmatrix} + c_2\begin{bmatrix} 6 \\ 4 \end{bmatrix} = \begin{bmatrix} 4 \\ -9 \end{bmatrix}$$

This breaks down into a broken algebraic system:
1) $3c_1 + 6c_2 = 4$
2) $2c_1 + 4c_2 = -9 \implies 3c_1 + 6c_2 = -13.5$

Because $3c_1 + 6c_2$ cannot simultaneously equal $4$ and $-13.5$, **no solution exists**. Geometrically, the coordinate $(4, -9)$ lies completely outside our 1D line universe.

---

To keep your geometric intuition completely clear, separate these two concepts as a **verb** versus a **noun**:

| Concept | The Intuition | What it represents | Grid Perspective |
| :--- | :--- | :--- | :--- |
| **Linear Transformation** | The **Verb** (The Action) | The dynamic machine or mapping rule that grabs *any* input vector from the universe and moves it to a new location. | We visualize the **transformed/rotated grid** to see how space warps dynamically. |
| **Column Space** | The **Noun** (The Destination) | The static geometric landing zone (a point, a line, or a plane) representing *every possible destination* that the transformation can reach. | We visualize it as a fixed subspace right on top of our **original, standard un-transformed grid** so we can easily see if a target vector $\mathbf{b}$ is reachable. |

### Matrix Inverse

When we transform a vector or a grid using a matrix, we change the coordinate system. The **Inverse Matrix** is the exact mathematical mechanism designed to reverse that change.

---
In basic arithmetic, the "undo" operation for addition brings us back to **zero** ($5 - 5 = 0$). However, matrix transformations are operations of **multiplication, scaling, and rotation**. In multiplication, the neutral baseline is **one** ($5 \times \frac{1}{5} = 1$).

In Linear Algebra, our version of "one" is the **Identity Matrix ($I$)**:
$$I = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}$$

The Identity Matrix represents the perfect, un-warped, standard 2D grid where $\hat{i} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ and $\hat{j} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$.

* **Matrix $A$:** Bends and warps the standard grid into a new shape.
* **Matrix $A^{-1}$ (The Inverse):** Grabs that already-warped grid and bends it precisely back into the standard Identity grid.

Therefore, applying a transformation and then immediately applying its inverse leaves the grid completely unchanged:
$$A^{-1}A = I$$

---
There is **only one single universe (the grid)**, and every transformation modifies its current state.

Let's apply a vertical shear matrix $A$:
$$A = \begin{bmatrix} 1 & 0 \\ 1 & 1 \end{bmatrix}$$

This matrix commands the grid to shift:
* $\hat{i}$ lands at the warped coordinate $\begin{bmatrix} 1 \\ 1 \end{bmatrix}$
* $\hat{j}$ stays at $\begin{bmatrix} 0 \\ 1 \end{bmatrix}$

The inverse matrix for this transformation is:
$$A^{-1} = \begin{bmatrix} 1 & 0 \\ -1 & 1 \end{bmatrix}$$

Its columns are $\begin{bmatrix} 1 \\ -1 \end{bmatrix}$ and $\begin{bmatrix} 0 \\ 1 \end{bmatrix}$. This matrix represents a rule: **"Take whatever grid you are given, and slide its right side DOWN by 1 unit."**

Because the transformations happen sequentially on the *same* grid, we don't apply $A^{-1}$ to a clean basis. We apply it directly to the warped vector $\begin{bmatrix} 1 \\ 1 \end{bmatrix}$ that matrix $A$ left behind.

$$A^{-1} (\text{Warped } \hat{i}) = \begin{bmatrix} 1 & 0 \\ -1 & 1 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix}$$

Computing this as a linear combination of $A^{-1}$'s columns using the warped coordinates as weights:
$$1 \begin{bmatrix} 1 \\ -1 \end{bmatrix} + 1 \begin{bmatrix} 0 \\ 1 \end{bmatrix} = \begin{bmatrix} 1 + 0 \\ -1 + 1 \end{bmatrix} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$$

**The grid is restored!** The down-shear instructions of the inverse perfectly counteracted the up-shear damage of the original matrix.

---


Now we can geometrically understand why a matrix with a **determinant of 0** cannot be inverted (it is "singular").

When $\det = 0$, the matrix squishes the entire 2D grid flat onto a 1D line or a 0D point (the Column Space). 

$$\begin{bmatrix} 2D \\ \text{Grid} \end{bmatrix} \xrightarrow{\quad A \quad} \begin{bmatrix} 1D \\ \text{Line} \end{bmatrix} \xrightarrow{\quad A^{-1} \text{ ???} \quad} \begin{bmatrix} 2D \\ \text{Grid} \end{bmatrix}$$

To build an inverse matrix $A^{-1}$ for a collapsed space, that matrix would have to take a flat 1D line and explode it back out into a full 2D plane. But the information was completely destroyed during the collapse—infinitely many points were crushed into the exact same locations. 

Because you cannot reconstruct a 2D space from a 1D line without losing track of where the points originally came from, **a collapsed space has no inverse.**

### Ax = v

**The Forward Mapping: $Ax = v$**

The equation $Ax = v$ describes the mathematical action of a linear transformation operator applied to an input vector. Breaking down the components geometrically:

* $x$ represents the initial vector residing in the standard, unaltered input space. Its coordinates dictate a specific position relative to the standard basis vectors $\hat{i}$ and $\hat{j}$.
* $A$ is the transformation matrix. Geometrically, it acts as an operator that warps the coordinate system, altering the grid lines based on where its column vectors land.
* $v$ is the resulting output vector. It represents the precise position where the original vector $x$ lands within the newly transformed environment.

When you multiply a matrix $A$ by a vector $x$, you are tracking how that specific coordinate position shifts when the underlying space undergoes a linear transformation. The output vector $v$ is simply the final destination coordinate.

**The Inverse Mapping: $x = A^{-1}v$**

The equation $x = A^{-1}v$ addresses the inverse problem: given a known final destination vector $v$ within a transformed space, how do we reconstruct its original starting coordinates $x$?

* $v$ represents the vector's coordinates in the altered output space.
* $A^{-1}$ is the inverse transformation matrix. It represents a precise geometric counter-action that reverses the exact rotations, scales, or shears executed by matrix $A$.
* $x$ is the restored vector back in its original, standard coordinate framework.

Because sequential linear transformations occur on the same continuous space, the operator $A^{-1}$ acts directly on the transformed vector $v$. It systematically un-rotates and un-shears the space, mapping the coordinates directly back to the standard identity grid.

**Algebraic Alignment with Geometry**

The algebraic derivation perfectly mirrors this physical restoration. If we begin with the forward transformation:
$$Ax = v$$

Applying the inverse operator $A^{-1}$ to both sides of the relation yields:
$$A^{-1}(Ax) = A^{-1}v$$

Because the consecutive application of a transformation and its exact inverse cancels out perfectly, the operator product reduces to the identity matrix $I$:
$$A^{-1}A = I \implies Ix = A^{-1}v$$

Since the identity matrix leaves any vector unchanged ($Ix = x$), the equation simplifies to isolate the original starting vector:
$$x = A^{-1}v$$

**The Constraint of Dimensional Collapse**

This geometric framework explains why a matrix with a determinant of zero ($\det(A) = 0$) cannot possess an inverse. A zero determinant signifies that the transformation $A$ has compressed the coordinate space into a lower dimension—collapsing a two-dimensional plane into a one-dimensional line or a single point. 

During this dimensional compression, critical coordinate information is permanently lost because infinitely many distinct input vectors are mapped onto the exact same output coordinates. Because an operator cannot reconstruct a higher-dimensional space from a lower-dimensional subspace without losing the tracking history of the original points, a collapsed space cannot be mapped backward. Thus, when $\det(A) = 0$, the inverse matrix $A^{-1}$ does not exist, and the equation $x = A^{-1}v$ cannot be solved uniquely.


### Row space


Matrix vector multiplication, are based on dot products and linear combination

The matrix-vector multiplication $A\vec{x} = \vec{b}$ represents a linear transformation mapping a vector from an input space $\mathbb{R}^n$ to an output space $\mathbb{R}^m$, where $A \in \mathbb{R}^{m \times n}$, $\vec{x} \in \mathbb{R}^n$, and $\vec{b} \in \mathbb{R}^m$. This fundamental operation is governed by two complementary perspectives: the row perspective, which relies on inner (dot) products within the input space, and the column perspective, which relies on linear combinations within the output space.

When evaluated from the row perspective, each individual coordinate $b_i$ of the output vector is computed as the dot product of the $i$-th row vector of the matrix, denoted as $\vec{r}_i$, and the input vector $\vec{x}$:
$$b_i = \vec{r}_i \cdot \vec{x}$$
Because both $\vec{r}_i$ and $\vec{x}$ are elements of $\mathbb{R}^n$, these dot products are computed entirely within the geometry of the input space. Each row functions as a directional operator that measures the structural alignment and projection of the input vector along that specific dimension.

When evaluated from the column perspective, the multiplication is executed as a linear combination of the column vectors $\vec{c}_j$ of the matrix, weighted by the scalar components $x_j$ of the input vector:
$$\vec{b} = x_1\vec{c}_1 + x_2\vec{c}_2 + \dots + x_n\vec{c}_n$$
Because each column vector $\vec{c}_j$ and the resulting vector $\vec{b}$ reside in $\mathbb{R}^m$, this linear combination exists entirely within the output space. The span of all possible linear combinations of these columns defines the column space $\mathcal{C}(A)$, representing the total reachable subspace within the output domain.

The row space $\mathcal{C}(A^T)$ is the subspace spanned by the row vectors of the matrix and lives strictly within the input space $\mathbb{R}^n$. It represents the precise subspace of input configurations that the matrix can actively detect and transform. 

The fundamental algebraic property of the row space is its strict orthogonal complementarity to the null space $\mathcal{N}(A)$. Any vector $\vec{x}_{\text{null}}$ belonging to the null space is mapped to the zero vector under transformation, which can be expressed through the row perspective as a system of vanishing dot products:
$$\begin{bmatrix} \vec{r}_1 \cdot \vec{x}_{\text{null}} \\ \vec{r}_2 \cdot \vec{x}_{\text{null}} \\ \vdots \\ \vec{r}_m \cdot \vec{x}_{\text{null}} \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ \vdots \\ 0 \end{bmatrix}$$
Because the dot product of every row vector with a null space vector is identically zero, every vector in the row space is perpendicular to every vector in the null space.

Consequently, any arbitrary input vector $\vec{x} \in \mathbb{R}^n$ can be uniquely decomposed into two orthogonal components: a row space component $\vec{x}_{\text{row}} \in \mathcal{C}(A^T)$ and a null space component $\vec{x}_{\text{null}} \in \mathcal{N}(A)$:
$$\vec{x} = \vec{x}_{\text{row}} + \vec{x}_{\text{null}}$$

Applying the linear transformation allows the matrix operator to distribute across these components:
$$A\vec{x} = A(\vec{x}_{\text{row}} + \vec{x}_{\text{null}}) = A\vec{x}_{\text{row}} + A\vec{x}_{\text{null}}$$

By definition, the null space component evaluates to zero, collapsing the expression:
$$A\vec{x} = A\vec{x}_{\text{row}}$$

This demonstrates that the matrix isolates and discards all coordinate information perpendicular to its row space. The row space contains the only input coordinates that survive the transformation. The matrix maps this row space component bijectively onto the column space in the output domain, preserving its exact dimensional structure. Because of this direct geometric mapping, the dimension of the row space always equals the dimension of the column space, establishing the fundamental rank of the matrix.

Every input vector $\vec{x} \in \mathbb{R}^n$ can be uniquely decomposed into two orthogonal components: a row space component $\vec{x}_{\text{row}}$ and a null space component $\vec{x}_{\text{null}}$. This relationship is defined by the fundamental orthogonal complementarity:
$$\vec{x} = \vec{x}_{\text{row}} + \vec{x}_{\text{null}}$$

Where $\vec{x}_{\text{row}} \in \mathcal{C}(A^T)$ and $\vec{x}_{\text{null}} \in \mathcal{N}(A)$. Geometrically, the row space and null space act as two perpendicular "sub-universes" that together span the entire input space.

When a matrix $A$ acts upon an input vector $\vec{x}$, the linearity of the transformation allows us to distribute the matrix across these components:
$$A\vec{x} = A(\vec{x}_{\text{row}} + \vec{x}_{\text{null}}) = A\vec{x}_{\text{row}} + A\vec{x}_{\text{null}}$$

By the definition of the null space, any vector $\vec{x}_{\text{null}}$ satisfies $A\vec{x}_{\text{null}} = \vec{0}$. Consequently, the transformation simplifies to:
$$A\vec{x} = A\vec{x}_{\text{row}}$$

The matrix effectively acts as a geometric filter:
* **The Filter:** The matrix ignores all information residing in the null space. If a matrix is rank-deficient (rank $< n$), it maps an entire subspace of the input (the null space) to the zero vector in the output space.
* **The Transformation:** Only the projection of the input vector onto the row space is preserved and mapped into the column space $\mathcal{C}(A)$. 

Because the matrix maps every vector in the null space to the origin, any two input vectors $\vec{x}_1$ and $\vec{x}_2$ that share the same row space component $\vec{x}_{\text{row}}$ will produce the identical output vector $\vec{b}$:
$$\text{If } \vec{x}_1 = \vec{x}_{\text{row}} + \vec{n}_1 \text{ and } \vec{x}_2 = \vec{x}_{\text{row}} + \vec{n}_2 \text{ (where } \vec{n}_1, \vec{n}_2 \in \mathcal{N}(A))$$
$$A\vec{x}_1 = A\vec{x}_{\text{row}} = A\vec{x}_2$$

This confirms your intuition: the column space is the set of all reachable output destinations, and the row space is the filter that determines which input configurations actually influence the transformation. When rank is low, the matrix "collapses" the dimensionality of the input space onto a smaller subspace in the output space, effectively creating a "many-to-one" mapping.

## Characteristic polynomial

The characteristic equation $\det(A - \lambda I) = 0$ identifies the specific scalar scaling factors ($\lambda$) that, when subtracted from a square matrix $A$, force the transformation to lose linear independence, drop rank, and collapse its tracking area or volume to zero. 

Consider the full-rank 2D transformation matrix $A$:

$$A = \begin{pmatrix} 1 & 2 \\ 2 & 1 \end{pmatrix}$$

To find the scalar shift that triggers a structural collapse, we subtract $\lambda$ from the diagonal components and set the determinant of the resulting matrix to zero:

$$\det(A - \lambda I) = \det \begin{pmatrix} 1 - \lambda & 2 \\ 2 & 1 - \lambda \end{pmatrix} = 0$$

Computing the 2x2 determinant ($(ad - bc)$) yields the characteristic polynomial equation:

$$(1 - \lambda)(1 - \lambda) - (2)(2) = 0$$
$$\lambda^2 - 2\lambda - 3 = 0$$

Factoring this quadratic equation as $(\lambda - 3)(\lambda + 1) = 0$ reveals the two rank-collapsing scalars: $\lambda = 3$ and $\lambda = -1$.

Plugging the scalar $\lambda = 3$ back into the shifted matrix demonstrates this geometric collapse directly:

$$A - 3I = \begin{pmatrix} 1 - 3 & 2 \\ 2 & 1 - 3 \end{pmatrix} = \begin{pmatrix} -2 & 2 \\ 2 & -2 \end{pmatrix}$$

The columns $\begin{pmatrix} -2 \\ 2 \end{pmatrix}$ and $\begin{pmatrix} 2 \\ -2 \end{pmatrix}$ are now scalar multiples of one another, making them linearly dependent. The modified transformation can no longer span a 2D plane; it flattens all of 2D space onto a single 1D line ($y = -x$). This structural collapse drops the matrix to Rank 1 and reduces its spatial volume (determinant) to exactly zero.

## Correlation and covariance

The dot product between two vectors $x$ and $y$ evaluates their unnormalized geometric alignment, defined algebraically as the sum of their element-wise products and geometrically as:

$$x \cdot y = \|x\| \|y\| \cos(\theta)$$

By normalizing the dot product using the magnitudes of the vectors, we derive the cosine similarity. This metric isolates the directional alignment (the angle $\theta$) between the vectors, stripping away the influence of their individual lengths:

$$\text{Cosine Similarity} = \cos(\theta) = \frac{x \cdot y}{\|x\| \|y\|}$$

The Pearson correlation coefficient measures the strength of the linear relationship between two variables. It is calculated by dividing the covariance of the variables by the product of their standard deviations:

$$r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2} \sqrt{\sum (y_i - \bar{y})^2}}$$

Fundamentally, Pearson correlation is mathematically equivalent to the cosine similarity of mean-centered vectors. If we transform our original vectors by subtracting their respective scalar means (creating centered vectors $\tilde{x} = x - \bar{x}$ and $\tilde{y} = y - \bar{y}$), the Pearson formula simplifies perfectly into the cosine similarity equation:

$$r = \frac{\tilde{x} \cdot \tilde{y}}{\|\tilde{x}\| \|\tilde{y}\|} = \cos(\theta_{\text{centered}})$$

Therefore, cosine similarity and Pearson correlation evaluate the exact same geometric alignment. Pearson correlation is simply the cosine similarity calculated after shifting the vectors so their means are centered at the origin.

If we let $X_c = X - \bar{X}$ and $Y_c = Y - \bar{Y}$ be the mean-centered vectors, their dot product is written as:
$$X_c \cdot Y_c = \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})$$

Covariance is the dot product of the mean-centered vectors divided by $n - 1$:
$$\text{Covariance}(X, Y) = \frac{X_c \cdot Y_c}{n - 1} = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{n - 1}$$

Pearson correlation divides the covariance by the product of the standard deviations, which cancels out the $n - 1$ denominators and simplifies directly to the Cosine Similarity of the mean-centered vectors:
$$r = \frac{\text{Covariance}(X,Y)}{\text{StdDev}(X) \times \text{StdDev}(Y)} = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^{n} (X_i - \bar{X})^2} \sqrt{\sum_{i=1}^{n} (Y_i - \bar{Y})^2}}$$

**Covariance** is a statistical measure that quantifies the directional relationship between two random variables. In plain terms, it tells you whether two variables tend to move in the same direction or in opposite directions.

---
When you calculate the covariance between Variable $X$ and Variable $Y$, the resulting number will fall into one of three categories:

1. **Positive Covariance ($> 0$):** Indicates that the two variables move together. When $X$ is above its average, $Y$ tends to be above its average as well (e.g., *Height* and *Weight*).
2. **Negative Covariance ($< 0$):** Indicates that the variables move in opposite directions. When $X$ goes up, $Y$ tends to go down (e.g., *Elevation* and *Temperature*).
3. **Zero Covariance ($\approx 0$):** Indicates that there is no linear relationship between the two variables; they move completely independently of each other (e.g., *Shoe Size* and *Intelligence*).

---

For a sample dataset, the covariance between two variables $X$ and $Y$ is calculated by finding the average product of their deviations from their respective means:

$$\text{Covariance}(X, Y) = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{n - 1}$$

Where:
* $X_i$ and $Y_i$ are individual data points.
* $\bar{X}$ and $\bar{Y}$ are the sample means of the variables.
* $n$ is the total number of data points.
* Covariance between two variables is the dot product between x_mean_centered and y_mean_centered
* Covariance takes two variables at a time, that is why matrix is needed for multiple variables

> ⚠️ **The Scale Limitation:** Covariance only tells you the **direction** of the relationship, not the strength. Because it is calculated using the raw units of your data, the final number can be massive or tiny purely based on whether you measure in kilometers or millimeters. To measure the actual strength, you must normalize covariance into **correlation**.

````python
#Creating covariance 
height = np.random.normal ( 170 , 5 ,100 )
weight = np.random.normal ( 70, 10, 100)
age = np.random.normal(35, 8, 100)
X = np.vstack([height, weight, age]).T
````

````python
X_centered = X - np.mean( X , axis = 0 )
n = len (X) 
cov_matrix = (X_centered.T @ X_centered) / (n - 1) #This takes the dot product between two observation (itself + to each other)
cov_matrix
````

````python
np.cov(X, rowvar=False)
````

This covariance matrix gives you a complete map of how your generated Height, Weight, and Age data behave. 

The diagonal values ($27.18$, $103.03$, and $72.86$) represent the **variances**. They tell you how much each individual variable spreads out from its own mean. For example, $27.18$ is the variance of Height itself. If you take its square root ($\sqrt{27.18} \approx 5.21$), you get a standard deviation very close to the $5\text{ cm}$ you originally specified.

The off-diagonal values represent the **covariances** between different pairs. Because these numbers (like $-0.47$ for Height vs. Weight, or $-8.62$ for Weight vs. Age) are negative, it tells you that in this random sample, the variables have a slight tendency to move in opposite directions. The matrix product $A^T A$ is fundamentally a Gram matrix that computes the dot product of every column of $A$ with every other column of $A$. For a matrix $A$ composed of column vectors $\begin{bmatrix} \mathbf{c}_1 & \mathbf{c}_2 & \dots & \mathbf{c}_n \end{bmatrix}$, the operation is explicitly represented as:

$$
A^T A = \begin{bmatrix} 
\leftarrow \mathbf{c}_1^T \rightarrow \\ 
\leftarrow \mathbf{c}_2^T \rightarrow \\ 
\vdots \\ 
\leftarrow \mathbf{c}_n^T \rightarrow 
\end{bmatrix}
\begin{bmatrix} 
\uparrow & \uparrow & & \uparrow \\ 
\mathbf{c}_1 & \mathbf{c}_2 & \dots & \mathbf{c}_n \\ 
\downarrow & \downarrow & & \downarrow 
\end{bmatrix} = 
\begin{bmatrix} 
\mathbf{c}_1 \cdot \mathbf{c}_1 & \mathbf{c}_1 \cdot \mathbf{c}_2 & \dots & \mathbf{c}_1 \cdot \mathbf{c}_n \\ 
\mathbf{c}_2 \cdot \mathbf{c}_1 & \mathbf{c}_2 \cdot \mathbf{c}_2 & \dots & \mathbf{c}_2 \cdot \mathbf{c}_n \\ 
\vdots & \vdots & \ddots & \vdots \\ 
\mathbf{c}_n \cdot \mathbf{c}_1 & \mathbf{c}_n \cdot \mathbf{c}_2 & \dots & \mathbf{c}_n \cdot \mathbf{c}_n 
\end{bmatrix}
$$

Geometrically, the diagonal elements $\mathbf{c}_i \cdot \mathbf{c}_i$ yield the squared Euclidean norm $\|\mathbf{c}_i\|^2$ of each column, measuring its structural scale and magnitude. The off-diagonal elements $\mathbf{c}_i \cdot \mathbf{c}_j$ evaluate the directional alignment between distinct vectors; a value of zero explicitly confirms that the columns are orthogonal.

Statistically, if the columns of matrix $A$ represent individual features that have been mean-centered, this geometric product transforms directly into a statistical profile. The diagonal entries map to the variance of each individual feature, quantifying data dispersion along that dimension. The off-diagonal entries represent the covariance between pairs of features, capturing how they linearly co-vary. Consequently, scaling this product by the degree of freedom, $\frac{1}{n-1} A^T A$, constructs the standard empirical covariance matrix utilized throughout machine learning applications.

- Recalling the orthogonal decompositon of two vectors first:

Given a target vector $\vec{v}$ and a baseline vector $\vec{u}$, our goal is to split $\vec{v}$ into a parallel component $\vec{v}_{\parallel}$ and a perpendicular component $\vec{v}_{\perp}$ relative to $\vec{u}$, satisfying the equation:
$$\vec{v} = \vec{v}_{\parallel} + \vec{v}_{\perp}$$

**1. The Parallel Condition**
Because $\vec{v}_{\parallel}$ points in the exact same direction as $\vec{u}$, it must be a scalar multiple of $\vec{u}$. We introduce an unknown scaling factor $c$:
$$\vec{v}_{\parallel} = c\vec{u}$$

**2. Defining the Perpendicular Component**
Using basic vector subtraction, we can define the perpendicular component as the original vector minus the parallel component we just defined:
$$\vec{v}_{\perp} = \vec{v} - \vec{v}_{\parallel} = \vec{v} - c\vec{u}$$

**3. The Perpendicular Condition (Dot Product)**
For $\vec{v}_{\perp}$ to be strictly perpendicular to the baseline vector $\vec{u}$, their dot product must equal zero:
$$\vec{v}_{\perp} \cdot \vec{u} = 0$$

Substituting our expression for $\vec{v}_{\perp}$ into this orthogonality condition gives:
$$(\vec{v} - c\vec{u}) \cdot \vec{u} = 0$$

**4. Solving for the Scalar $c$**
Distributing the dot product across the terms inside the parentheses:
$$(\vec{v} \cdot \vec{u}) - c(\vec{u} \cdot \vec{u}) = 0$$

Moving the negative term to the right side:
$$\vec{v} \cdot \vec{u} = c(\vec{u} \cdot \vec{u})$$

Isolating the unknown scalar $c$ by dividing both sides by $(\vec{u} \cdot \vec{u})$:
$$c = \frac{\vec{v} \cdot \vec{u}}{\vec{u} \cdot \vec{u}}$$

**5. Final Component Formulas**
Plugging the calculated value of $c$ back into our original definitions yields the full formulas for orthogonal decomposition:

* **Parallel Component (Projection):**
$$\vec{v}_{\parallel} = \left( \frac{\vec{v} \cdot \vec{u}}{\vec{u} \cdot \vec{u}} \right) \vec{u}$$

* **Perpendicular Component (Error/Leftover):**
$$\vec{v}_{\perp} = \vec{v} - \left( \frac{\vec{v} \cdot \vec{u}}{\vec{u} \cdot \vec{u}} \right) \vec{u}$$


- This is the same concept in the left inverse which gives approximation by projection vector 'v' in the column space


 **Moore-Penrose Pseudoinverse** (often just called the pseudoinverse).

When you have more equations than unknowns (a "tall" matrix), the system is overdetermined, meaning a perfect solution usually doesn't exist. The Moore-Penrose theorem proves that there is still a unique, optimal "best fit" solution that minimizes the error. For a tall matrix with linearly independent columns, that pseudoinverse is exactly the **left inverse** we are about to derive.

Let's use the exact same 5-step logic you just mastered for a single vector, but upgrade it to handle a matrix $A$ representing a whole subspace (like a 2D plane in 3D space) and a target vector $\vec{v}$.

---
The Matrix Derivation

Given a target vector $\vec{v}$ and a tall matrix $A$, our goal is to split $\vec{v}$ into a parallel component $\hat{v}$ (which lies flat in the column space of $A$) and a perpendicular error component $\vec{e}$:


$$\vec{v} = \hat{v} + \vec{e}$$

1. The Parallel Condition

Because the projection $\hat{v}$ must live entirely inside the space spanned by the columns of $A$, it must be a linear combination of those columns.

Instead of multiplying by a single scalar $c$ (like we did for one vector), we multiply the matrix $A$ by a **vector of weights** $\vec{x}$:


$$\hat{v} = A\vec{x}$$

Our goal now is to find this unknown vector $\vec{x}$.

2. Defining the Perpendicular Component

Just like before, the error vector $\vec{e}$ is simply whatever is left over when you subtract the parallel piece from the original vector:


$$\vec{e} = \vec{v} - \hat{v} = \vec{v} - A\vec{x}$$

3. The Perpendicular Condition (Matrix Dot Product)

For $\vec{e}$ to be perfectly perpendicular to the entire subspace, it must be perpendicular to **every single column** of $A$ individually.

If we transpose $A$ to turn its columns into rows ($A^T$), multiplying $A^T$ by $\vec{e}$ computes the dot product of $\vec{e}$ with every column at the same time. For all of them to be perpendicular ($90^\circ$), this matrix product must equal the zero vector:


$$A^T \vec{e} = \vec{0}$$

Substituting our definition of $\vec{e}$ into this condition gives:


$$A^T (\vec{v} - A\vec{x}) = \vec{0}$$

4. Solving for the Vector $\vec{x}$

Now we use matrix algebra to isolate $\vec{x}$.

Distribute $A^T$ across the terms inside the parentheses:


$$A^T \vec{v} - A^T A \vec{x} = \vec{0}$$

Move the negative term to the right side to get the famous **Normal Equations**:


$$A^T \vec{v} = A^T A \vec{x}$$

Because $A$ is a tall matrix with linearly independent columns, the matrix $A^T A$ is a perfectly square, symmetric, and invertible matrix. To isolate $\vec{x}$, multiply both sides from the **left** by the inverse $(A^T A)^{-1}$:


$$\vec{x} = (A^T A)^{-1} A^T \vec{v}$$

5. Final Matrix Formulas

Now we can write out the two final components.

* **The Left Inverse (Moore-Penrose Pseudoinverse):**
The machinery that converts the floating vector $\vec{v}$ into the optimal weight vector $\vec{x}$ is:

$$A_{\text{left}}^{-1} = (A^T A)^{-1} A^T$$


* **The Projection Matrix:**
To find the actual spatial coordinates of the projection $\hat{v}$ on the plane, plug $\vec{x}$ back into our first step ($\hat{v} = A\vec{x}$):

$$\hat{v} = A (A^T A)^{-1} A^T \vec{v}$$



---

Notice the Beautiful Symmetry:

Look at how the single-vector scalar $c$ and the matrix weight vector $\vec{x}$ mirror each other perfectly:

| Component | Single Vector Geometry | Subspace Matrix Geometry |
| --- | --- | --- |
| **The "Multiplier"** | $c = \frac{\vec{u} \cdot \vec{v}}{\vec{u} \cdot \vec{u}} = (\vec{u}\cdot\vec{u})^{-1}(\vec{u}\cdot\vec{v})$ | $\vec{x} = (A^T A)^{-1} A^T \vec{v}$ |
| **The Final Projection** | $\vec{v}_{\parallel} = \vec{u} c$ | $\hat{v} = A \vec{x}$ |

Your target vector $\vec{v}$ lives in the output space. When we decompose it into parallel and perpendicular pieces, we are splitting the output space into two orthogonal halves:

1. **The Parallel Piece ($\hat{v}$):** Lives in the **Column Space** of $A$ (denoted as $C(A)$).
2. **The Perpendicular Piece ($\vec{e}$):** Lives in the **Left Null Space** of $A$ (denoted as $N(A^T)$).

Why is it called the Left Null Space?

Think back to the "90-degree tester" equation we just derived in the previous step:


$$A^T \vec{e} = \vec{0}$$

By literal definition, the "Null Space" of any matrix is the collection of all vectors that get multiplied by that matrix and crushed down to zero. Because $\vec{e}$ gets multiplied by $A^T$ and crushed to zero, **$\vec{e}$ is by definition in the Null Space of $A^T$**.

We call it the *Left* Null Space simply because if you flip the vectors to rows, $\vec{e}^T$ sits on the left side of $A$ ($\vec{e}^T A = \vec{0}^T$).

---

Does the Perpendicular Component Get "Ignored"?

**Yes, absolutely! Wiped out completely.** Your intuition here is flawless.

When you feed the target vector $\vec{v}$ into the left inverse machinery to find your best guess $\vec{x}$, the math explicitly ignores the perpendicular error. Watch how beautifully it collapses:

Remember that $\vec{v} = \hat{v} + \vec{e}$ (Parallel + Perpendicular). Let's multiply it by the Left Inverse:

$$\vec{x} = (A^T A)^{-1} A^T \vec{v}$$

$$\vec{x} = (A^T A)^{-1} A^T (\hat{v} + \vec{e})$$

Now, let's distribute that matrix across the addition:

$$\vec{x} = (A^T A)^{-1} A^T \hat{v} + (A^T A)^{-1} \underbrace{A^T \vec{e}}_{\vec{0}}$$

Because $\vec{e}$ lives in the Left Null Space, $A^T \vec{e}$ becomes exactly $\vec{0}$. The entire right side of the equation vanishes:

$$\vec{x} = (A^T A)^{-1} A^T \hat{v} + \vec{0}$$

The Left Inverse machine completely filters out the perpendicular noise ($\vec{e}$) and only calculates the answer based on the clean, parallel shadow ($\hat{v}$) that actually fits inside the column space.

| Space | What lives there? | The Subspace Split ($90^\circ$ Complements) |
| --- | --- | --- |

| **Output Space** ($\mathbb{R}^m$) | Target vectors ($\vec{v}$) | **Column Space** $\perp$ **Left Null Space** ($A^T\vec{e}=\vec{0}$) |

Here is the complete, detailed derivation from absolute scratch.

---
The Input Space Split

Imagine a wide $2 \times 3$ matrix $A$. It maps a 3D input vector $\vec{x}$ to a 2D output vector $\vec{v}$. Because you are crushing 3 dimensions down to 2, there is an entire line of 3D input vectors that all map to the exact same output vector $\vec{v}$.

This line of infinite solutions is shifted away from the origin. We want to find the **single unique vector $\vec{x}$ on this line that is closest to the origin** (the shortest vector).

Every possible input vector $\vec{x}$ can be broken down into two orthogonal components inside the input space:


$$\vec{x} = \vec{x}_{\text{row}} + \vec{x}_{\text{null}}$$

* **$\vec{x}_{\text{null}}$ lives in the Null Space:** By definition, $A\vec{x}_{\text{null}} = \vec{0}$. This component adds length to the vector but contributes absolutely nothing to reaching the target $\vec{v}$.
* **$\vec{x}_{\text{row}}$ lives in the Row Space:** This is the component that actually does the work to hit the target $\vec{v}$.

To make $\vec{x}$ as short as possible, the path from the origin to the line of solutions must hit at a perfect $90^\circ$ angle. That means the shortest vector must have **zero null space component** ($\vec{x}_{\text{null}} = \vec{0}$).

> **Our Geometric Constraint:** The optimal, shortest solution $\vec{x}$ must live **entirely inside the Row Space** of $A$.

---

Derivation

Now let's turn this geometric constraint into pure matrix algebra. We want to solve the system:


$$A\vec{x} = \vec{v}$$

Step 1: Express $\vec{x}$ as a Row Space Vector

The Row Space of $A$ is spanned by the rows of $A$. If we take the transpose matrix $A^T$, the rows of $A$ become the columns of $A^T$.

Therefore, saying *"$\vec{x}$ must live entirely in the Row Space"* is mathematically identical to saying *"$\vec{x}$ must be a linear combination of the columns of $A^T$."* We can write this as:


$$\vec{x} = A^T \vec{y}$$

Here, $\vec{y}$ is an unknown vector of weights that we need to discover.

Step 2: Substitute into the System Equation

We need our vector $\vec{x}$ to be a valid solution to our original equation $A\vec{x} = \vec{v}$. Let's substitute our Row Space definition ($\vec{x} = A^T \vec{y}$) directly into that equation:


$$A(A^T \vec{y}) = \vec{v}$$

Group the matrices together:


$$(AA^T)\vec{y} = \vec{v}$$

Step 3: Analyze the New Core Matrix ($AA^T$)

Let's look at the dimensions of $AA^T$:

* $A$ is a wide $2 \times 3$ matrix.
* $A^T$ is a tall $3 \times 2$ matrix.
* Multiplying a $2 \times 3$ by a $3 \times 2$ results in a **square $2 \times 2$ matrix**.

Because the rows of $A$ are linearly independent (which is true for a full-row-rank wide matrix), this small square matrix $AA^T$ is guaranteed to be fully invertible.

 Step 4: Solve for the Weight Vector $\vec{y}$

To isolate our mystery weight vector $\vec{y}$, we multiply both sides of the equation from the **left** by the inverse matrix $(AA^T)^{-1}$:


$$(AA^T)^{-1}(AA^T)\vec{y} = (AA^T)^{-1}\vec{v}$$

Since a matrix multiplied by its inverse is just the Identity matrix ($I$), it simplifies to:


$$\vec{y} = (AA^T)^{-1}\vec{v}$$

Step 5: Discover the Shortest Input Vector $\vec{x}$

Now that we know exactly what the weight vector $\vec{y}$ is, we plug it right back into our original definition of $\vec{x}$ from Step 1 ($\vec{x} = A^T \vec{y}$):

$$\vec{x} = A^T \left[ (AA^T)^{-1}\vec{v} \right]$$

Because matrix multiplication is associative, we can group the matrix parts together:


$$\vec{x} = \underbrace{A^T(AA^T)^{-1}}_{\text{Right Inverse}} \vec{v}$$

---
Why Is It Called the "Right" Inverse?

Look at what happens if we plug this solution back into our original equation $A\vec{x}$:

$$A\vec{x} = A \left[ A^T(AA^T)^{-1}\vec{v} \right]$$

$$A\vec{x} = \underbrace{(AA^T)(AA^T)^{-1}}_{I} \vec{v}$$

$$A\vec{x} = \vec{v}$$

The block of math $A^T(AA^T)^{-1}$ acts as a single operator. Because it sits on the **right side** of the matrix $A$ to cancel it out ($A \cdot A_{\text{right}}^{-1} = I$), it is crowned the **Right Inverse**.

You didn't drop an anchor onto a column space plane this time; instead, you used the $90^\circ$ orthogonality between the Row Space and the Null Space inside the input domain to strip away all the infinite, redundant choices, leaving you with the single, cleanest, shortest vector possible.

### Gram schimdit

Transforms the non orthogonal matrix in to  the orthogonal matrix using orthogonal vector decomposition

The Gram-Schmidt process transforms a matrix with linearly independent, non-orthogonal columns into an orthogonal matrix $Q$. Geometrically, this algorithm systematically removes the directional overlap between vectors. By stripping away these overlapping components, it leaves behind a set of vectors that are perfectly perpendicular ($90^\circ$ angles) and normalized to unit length ($1$).

When two vectors are not orthogonal, one vector casts a geometric "shadow" onto the other. In linear algebra, this shadow is called a **projection**. The fundamental trick of Gram-Schmidt is **orthogonal vector decomposition**: if you take a vector and subtract its projection (shadow) onto a subspace, the remaining vector (the residual) is guaranteed to be orthogonal to that subspace.

Consider a matrix $A$ with three non-orthogonal columns: $v_1$, $v_2$, and $v_3$. The process constructs the orthonormal columns $q_1$, $q_2$, and $q_3$ sequentially:

The first column $v_1$ is chosen as the reference vector. Since there are no prior vectors to overlap with, it does not need to be straightened. It is simply divided by its own length (norm) to make it a unit vector:
$$u_1 = v_1$$
$$q_1 = \frac{u_1}{\|u_1\|}$$

The second column $v_2$ likely points partially in the direction of $q_1$. To fix this, calculate the projection of $v_2$ onto $q_1$ and subtract it. This isolates the component of $v_2$ that is purely perpendicular to $q_1$. Finally, scale it to unit length:
$$u_2 = v_2 - \langle v_2, q_1 \rangle q_1$$
$$q_2 = \frac{u_2}{\|u_2\|}$$

The third column $v_3$ may lean toward both $q_1$ and $q_2$ . To ensure absolute independence, calculate its projection onto $q_1$ and its projection onto $q_2$(project the shadown on the plane formed by q1 and q2). Subtract both shadows simultaneously. The remaining vector is locked at a $90^\circ$ angle to both previous axes. Scale it to finish:
$$u_3 = v_3 - \langle v_3, q_1 \rangle q_1 - \langle v_3, q_2 \rangle q_2$$
$$q_3 = \frac{u_3}{\|u_3\|}$$

Packing these refined vectors back together yields the matrix $Q = [q_1 \ \ q_2 \ \ q_3]$. Because every overlapping shadow was subtracted out, the dot product of any two distinct columns is exactly $0$, and the dot product of any column with itself is $1$. The messy, skewed input matrix has been converted into a rigid, orthogonal transformation matrix.

````python
#Computing the gram schimidt
u1 = A [:,0]
u2 = A[ :,1] 
q1 = u1 / (lg.norm(u1))
q2 = u2 - ((q1.T @ u2) /( q1.T @ q1) ) * q1
q2 = q2/ lg.norm(q2)
Q = np.vstack( (q1,q2)).T
Q
#in this code, the denominator in projection formula is not required because q.q = 1 , it is already normalized before
````

3×3 Example

$$A = \begin{bmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \\ 1 & 0 & 1 \end{bmatrix}$$

Columns are: $a_1, a_2, a_3$

After GS we get Q with columns $q_1, q_2, q_3$ — all perpendicular unit vectors.

---

Compute R = Q.T @ A

Every entry $R_{ij} = q_i \cdot a_j$

$$R = \begin{bmatrix} q_1 \cdot a_1 & q_1 \cdot a_2 & q_1 \cdot a_3 \\ q_2 \cdot a_1 & q_2 \cdot a_2 & q_2 \cdot a_3 \\ q_3 \cdot a_1 & q_3 \cdot a_2 & q_3 \cdot a_3 \end{bmatrix}$$

Now the question is — **why are the lower entries zero?**

---

Focus on R[1,0] — Why is q2 · a1 = 0?

During GS, you built $q_2$ by:

$$q_2 = \text{normalize}(a_2 - (q_1 \cdot a_2)q_1)$$

You subtracted everything parallel to $q_1$ from $a_2$, then normalized.

So $q_2$ is **perfectly perpendicular to $q_1$ by construction.**

But wait — $a_1$ was used to **build** $q_1$. It points in the $q_1$ direction.

So:

$$q_2 \cdot a_1 = 0$$

Because $q_2 \perp q_1$ and $a_1$ lives entirely in the $q_1$ direction.

---

Focus on R[2,0] — Why is q3 · a1 = 0?

Same logic. During GS:

$$q_3 = \text{normalize}(a_3 - (q_1 \cdot a_3)q_1 - (q_2 \cdot a_3)q_2)$$

$q_3$ was built by removing ALL components of $q_1$ and $q_2$.

So $q_3 \perp q_1$, and since $a_1$ lives in $q_1$ direction:

$$q_3 \cdot a_1 = 0$$

---

Focus on R[2,1] — Why is q3 · a2 = 0?

$a_2$ lives in the span of $q_1$ and $q_2$ — that's exactly the space GS built from $a_1$ and $a_2$.

$q_3$ was built by removing ALL of $q_1$ and $q_2$ from $a_3$.

So $q_3$ is perpendicular to both $q_1$ and $q_2$ — meaning perpendicular to anything in their span — including $a_2$:

$$q_3 \cdot a_2 = 0$$

---

The Pattern

```
q2 was built after a1  →  q2 ⊥ a1  →  R[1,0] = 0
q3 was built after a1  →  q3 ⊥ a1  →  R[2,0] = 0
q3 was built after a2  →  q3 ⊥ a2  →  R[2,1] = 0
```

Every entry **below the diagonal** involves a later q dot an earlier a — and later q's are always perpendicular to earlier a's **by construction of GS.**

---

The Result

$$R = \begin{bmatrix} q_1 \cdot a_1 & q_1 \cdot a_2 & q_1 \cdot a_3 \\ 0 & q_2 \cdot a_2 & q_2 \cdot a_3 \\ 0 & 0 & q_3 \cdot a_3 \end{bmatrix}$$

Upper triangular — not by coincidence, but because **GS deliberately makes later q's perpendicular to all earlier a's.**

---

One Line

> R is upper triangular because GS builds each new q to be perpendicular to all previous a's — so every dot product below the diagonal is exactly zero.

## Eigen values and vector

The vectors that remains in their span after the linear transformation is called eigen vector.

A single vector has 1d span, so whole span is eigen vector if that span remains in the span and does not get rotated after the transformation


The equation can be computed if input vector remains same vector after transformation (just scaled) 

$$A\vec{v} = \lambda\vec{v}$$

$$A\vec{v} - \lambda\vec{v} = \mathbf{0}$$

$$A\vec{v} - \lambda I\vec{v} = \mathbf{0}$$

$$(A - \lambda I)\vec{v} = \mathbf{0}$$

$$\vec{v} \neq \mathbf{0} \implies \det(A - \lambda I) = 0$$

Here , the vector can be zero (trivial) but we are looking for trivial solution 

in which case we are looking for shfiitng the matrix by scalar which reduces the rank , the scalar which shifts the matrix to lower rank is eigen value

An $m \times m$ square matrix $A$ has $m$ eigenvalues ($\lambda_1, \lambda_2, \dots, \lambda_m$) and $m$ corresponding eigenvectors ($v_1, v_2, \dots, v_m$). This relationship gives us a system of $m$ individual equations:

$$Av_1 = \lambda_1 v_1$$

$$Av_2 = \lambda_2 v_2$$

$$\vdots$$

$$Av_m = \lambda_m v_m$$

Instead of dealing with $m$ separate equations, we can look at the system through a single, unified matrix equation. Let $P$ be the matrix whose columns are the eigenvectors of $A$:

$$P = \begin{bmatrix} | & | & & | \\ v_1 & v_2 & \dots & v_m \\ | & | & & | \end{bmatrix}$$

Next, we construct a diagonal matrix $D$ where the main diagonal entries are the corresponding eigenvalues, and all off-diagonal entries are zero:

$$D = \begin{bmatrix} \lambda_1 & 0 & \dots & 0 \\ 0 & \lambda_2 & \dots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \dots & \lambda_m \end{bmatrix}$$

If we multiply matrix $A$ by our eigenvector matrix $P$, the multiplication applies $A$ to each column vector individually:

$$AP = A \begin{bmatrix} v_1 & v_2 & \dots & v_m \end{bmatrix} = \begin{bmatrix} Av_1 & Av_2 & \dots & Av_m \end{bmatrix}$$

Substituting our initial eigenvalue equations into this matrix, we get:

$$AP = \begin{bmatrix} \lambda_1 v_1 & \lambda_2 v_2 & \dots & \lambda_m v_m \end{bmatrix}$$

We can achieve this exact same column-scaling effect by post-multiplying the eigenvector matrix $P$ by our diagonal eigenvalue matrix $D$:

$$PD = \begin{bmatrix} v_1 & v_2 & \dots & v_m \end{bmatrix} \begin{bmatrix} \lambda_1 & 0 & \dots & 0 \\ 0 & \lambda_2 & \dots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \dots & \lambda_m \end{bmatrix} = \begin{bmatrix} \lambda_1 v_1 & \lambda_2 v_2 & \dots & \lambda_m v_m \end{bmatrix}$$

Because both sides yield the identical result, we can compactly express the entire system as:

$$AP = PD$$

Assuming the eigenvectors are linearly independent, the matrix $P$ is invertible. We can isolate our diagonal matrix $D$ by multiplying both sides from the left by $P^{-1}$:

$$P^{-1}AP = P^{-1}PD$$

$$P^{-1}AP = D$$

Geometrically, this formula represents a powerful change of basis. The matrix $P$ contains our new basis vectors (the eigenbasis) expressed in terms of our standard coordinate system. Because $P$ moves vectors from the eigenbasis to the standard basis, its inverse, $P^{-1}$, does the opposite: it maps vectors from our standard coordinate system into the eigenbasis coordinate system.

When we evaluate the expression $P^{-1}AP$ from right to left, it reveals a three-step process: we take a vector, transform it using the complex matrix $A$ in standard coordinates, and then use $P^{-1}$ to translate that result into the eigenbasis. The derivation proves that this entire sequence is perfectly equivalent to evaluating $D$ directly—meaning that inside the coordinate system defined by the eigenvectors, the matrix transformation loses its complexity and simplifies into basic independent stretching along each axis.
