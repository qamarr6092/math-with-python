# Statistics

---

Data is a systematic collection of observations, measurements, or facts.

---


All data falls into one of two major categories:

A. Qualitative (Categorical) Data

Data that describes non-numerical qualities or characteristics. It cannot be measured on a numerical scale.

* **Nominal Data:** Categories with **no natural order or ranking**.
* *Examples:* Eye color (Blue, Brown), Blood type (A, B, O), Country.


* **Ordinal Data:** Categories with a **logical order or ranking**, but the mathematical distance between the ranks is not equal or measurable.
* *Examples:* Satisfaction level (Poor, Satisfactory, Excellent), Education level (High School, Bachelors, Masters).



B. Quantitative (Numerical) Data

Data expressed in numbers where mathematical calculations (like averages or subtraction) are valid.

* **Discrete Data:** Countable values that can only take **specific, distinct numbers** (usually whole numbers/integers).
* *Examples:* Number of siblings, pages in a book, cars in a parking lot. (You cannot have 2.5 siblings).


* **Continuous Data:** Measurable values that can take **any real number within a range**, including fractions and decimals.
* *Examples:* Height, weight, temperature, time.



---
3. The Four Levels of Measurement (NOIR)

For numerical data, statisticians use a scale of four levels (from lowest mathematical information to highest):

1. **Nominal:** Categories only. (e.g., Zip codes—they are numbers, but only used as labels).
2. **Ordinal:** Ordered categories. (e.g., Class rank: 1st, 2nd, 3rd).
3. **Interval:** Numerical data with **equal intervals between values, but no "true zero"**. Zero is just an arbitrary marker; negative numbers are possible.
* *Example:* Temperature in Celsius. $0^\circ\text{C}$ does not mean "no temperature at all."


4. **Ratio:** Numerical data with **equal intervals and a "true zero"** (where zero means complete absence of the quantity). Negative values are not possible.
* *Example:* Weight or Income. $0 means literally no money.

### Frequency Distributions

```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy as sc

height = np.array ( 
    [70.5, 67.2, 72.1, 68.9, 65.4, 71.3, 69.8, 74.2, 68.1, 70.0,
 66.7, 73.5, 69.1, 68.5, 71.9, 67.8, 70.4, 66.0, 72.8, 69.5,
 65.9, 71.1, 68.3, 70.6, 73.1, 67.4, 69.0, 75.0, 68.7, 70.2]
) 


height

```

```python
rounded_heights = np.round(height).astype(int)

x_coords = []
y_coords = []
counts = {}

for value in rounded_heights:
    if value not in counts:
        counts[value] = 1
    else:
        counts[value] += 1
    
    x_coords.append(value)
    y_coords.append(counts[value]) 
    
plt.figure(figsize=(10, 3))


plt.scatter(x_coords, y_coords, color='black', s=30)

plt.gca().get_yaxis().set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlim(64, 76)
plt.xticks(range(64, 77))
plt.ylim(0, max(y_coords) + 1)
plt.xlabel("Height (Inches)")
plt.tight_layout()
plt.show()
```

Because the raw data is just numbers we count how many times it appears for a range of class interval 

```python
# Define the boundaries of our class intervals
# (We use 75.1 to ensure the maximum value of 75.0 is included in the last interval)


bins = [65.0, 67.0, 69.0, 71.0, 73.0, 75.1]

# Calculate frequencies and boundaries
frequencies, bin_edges = np.histogram(height, bins=bins)


for i in range(len(frequencies)):
    start = bin_edges[i]
    # Represent the intervals as non-overlapping ranges (e.g., 65.0 - 66.9)
    end = bin_edges[i+1] - 0.1 if i < len(frequencies) - 1 else 75.0
    print(f"{start:.1f} - {end:.1f}{'':<6} | {frequencies[i]}")
```

Relative frequency is the fraction or proportion of times a specific value or class interval occurs relative to the total number of observations in a dataset.

While regular frequency tells you the raw count, relative frequency tells you what part of the whole that count represents.m

The Formula$$\text{Relative Frequency} = \frac{f}{n}$$Where:$f = \text{the frequency of a specific value or class interval}$

$n = \text{the total number of observations in the dataset}$

```python
print(f'Total Frequency : {np.sum(frequencies)} ')   #same to number of  obserations
```

```python
relative_freq = [ frequencies[i]/np.sum(frequencies) for i in range(len(frequencies))]
np.round(relative_freq,3)
```

So the class interval of 67- 68.9 inches appears around 26.7% in the dataset

```python

labels = ['65.0 - 66.9', '67.0 - 68.9', '69.0 - 70.9', '71.0 - 72.9', '73.0 - 75.0']


plt.figure(figsize=(6, 6))


plt.pie(
    relative_freq, 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=90, 
    wedgeprops={'edgecolor': 'black'}
)

# 4. Add title and display
plt.title("Relative Frequency of Heights")
plt.tight_layout()
plt.show()
```

This shows which height class interval appears most in terms of percentage in the dataset

```python

# 2. Convert decimal relative frequencies to percentages
percentages = [val * 100 for val in relative_freq]


plt.figure(figsize=(8, 5))
plt.bar(labels, percentages, color='skyblue', edgecolor='black')


plt.title("Relative Frequency of Heights")
plt.xlabel("Height Intervals (Inches)")
plt.ylabel("Percentage (%)")
plt.ylim(0, 35)  # Sets the Y-axis range from 0 to 35%

for i, val in enumerate(percentages):
    plt.text(i, val + 0.5, f"{val:.1f}%", ha='center')

plt.tight_layout()
plt.show()
```

1. **Type of Data**
* **Bar Chart:** Used for **qualitative (categorical)** data (e.g., fruit types, blood groups, country).
* **Histogram:** Used for **quantitative (continuous numerical)** data (e.g., height, weight, time).


2. **Gaps Between Bars**
* **Bar Chart:** Has **visible gaps** between the bars. This shows that the categories are separate and independent.
* **Histogram:** Has **no gaps** between the bars. The bars touch because they represent a continuous flow of number intervals (e.g., 65–67 inches right into 67–69 inches).


3. **The X-Axis (Horizontal Axis)**
* **Bar Chart:** The X-axis contains text labels for categories. You can rearrange the order of the bars (e.g., putting "Banana" before "Apple") without changing the meaning of the data.
* **Histogram:** The X-axis is a continuous number line. You **cannot** rearrange the bars because numbers must stay in strict chronological/numerical order.

Yes, when you pass a single integer like `bins=3`, **Python chooses the class intervals entirely on its own.** It does this using a simple, standard mathematical formula under the hood.

---


To make the intervals perfectly equal, Python runs three quick steps:

1. **Finds the Extremes:** It looks through your dataset to find the absolute minimum value and the absolute maximum value.
2. **Calculates the Total Range:** It subtracts the minimum from the maximum.
3. **Divides by $k$:** It divides that range by the number of bins ($k$) you requested to find the exact **class width** ($W$).

$$W = \frac{\text{Maximum} - \text{Minimum}}{k}$$

---


In your specific height dataset, Python scans the numbers and finds:

* **Minimum value:** $65.4$
* **Maximum value:** $75.0$

If you type `bins=3`, Python plugs those numbers right into the formula:

$$W = \frac{75.0 - 65.4}{3} = \frac{9.6}{3} = 3.2$$

Because the calculated class width is **3.2**, Python automatically builds these 3 seamless intervals starting exactly at your minimum value:

* **Interval 1:** $65.4 \text{ to } 68.6$
* **Interval 2:** $68.6 \text{ to } 71.8$
* **Interval 3:** $71.8 \text{ to } 75.0$

#### Cumulative Relative Frequency





While regular cumulative frequency tells you the running total of *counts* (e.g., "21 people"), cumulative relative frequency tells you the running total of *percentages* (e.g., "70% of people").



---

(highest points).

A distribution is unimodal if it has one peak, bimodal if it has two
peaks, and multimodal if it has three or more peaks

#### Symmetry


Another important consideration when examining the shape of a distribution is symmetry. A distribution that can be divided into two pieces that are mirror images of one
another is called symmetric.

Example : Bell shaped, uniform distribution

and both of them are unimodal

#### Skewness

A unimodal distribution that is not symmetric is either right skewed or left skewed.
A right-skewed distribution rises to its peak rapidly and comes back toward the horizontal axis more slowly—its “right tail” is longer than its “left tail.” A left-skewed
distribution rises to its peak slowly and comes back toward the horizontal axis more
rapidly—its “left tail” is longer than its “right tail.

A multimodal distribution can be skewed. Skewness is not limited to unimodal curves.

Why Skewness Applies to Multimodal Data

Skewness is simply a measure of asymmetry (how lopsided a graph is). It doesn't care how many hills (peaks) are inside your graph; 

it only cares if the left side and the right side of the entire distribution fail to mirror each other.

```python

intervals = pd.Series(pd.cut(male_height, bins=10))

freq_table = pd.DataFrame()


freq_table['Height Interval (cm)'] = intervals.value_counts().sort_index().index


freq_table['Midpoint'] = freq_table['Height Interval (cm)'].apply(lambda x: x.mid)


freq_table['Frequency'] = intervals.value_counts().sort_index().values
freq_table['Relative Frequency'] = intervals.value_counts(normalize=True).sort_index().values
freq_table['Percentage (%)'] = freq_table['Relative Frequency'] * 100


freq_table['Cumulative Frequency'] = freq_table['Frequency'].cumsum()
freq_table['Cumulative Relative Freq'] = freq_table['Relative Frequency'].cumsum()


freq_table
```

####  Median

Another frequently used measure of center is the median. Essentially, the median of a
data set is the number that divides the bottom 50% of the data from the top 50%.


Median of a Data Set
Arrange the data in increasing order.

 If the number of observations is odd, then the median is the observation
exactly in the middle of the ordered list.

 If the number of observations is even, then the median is the mean of the
two middle observations in the ordered list.

In both cases, if we let n denote the number of observations, then the median
is at position (n + 1)/2 in the ordered list

```python
data = np.sort(weights) 
for i in range(len(data) ) :
    print(f'{data[i]:.2f} kg')
```

Because the observations are even the median is 

```python
print(f'The median position = { (len(data) + 1) /2 } ')

print(f'The median of data is = {((data[4] + data[5]) / 2):.2f}kg')
```

```python
np.median( weights)
```

The Mean cares about **MAGNITUDE**: It adds up the actual values of every single data point. It is the center of gravity. If one data point grows larger, the mean must move to balance it.

The Median cares about **RANK (Position)**: It only cares about the order of the points. It is the geographic center—the point that splits your data into a 50/50 split. If the largest value in your dataset becomes 10 times larger, the median doesn't move a single millimeter because the order of the points hasn't changed.
