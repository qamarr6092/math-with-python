# Statistics

---

Data is a systematic collection of observations, measurements, or facts.

---


All data falls into one of two major categories:

A. Qualitative (Categorical) Data

Data that describes non-numerical qualities or characteristics. It cannot be measured on a numerical scale.

* **Nominal Data:** Categories with **no natural order or ranking**.
* *Examples:* Eye color (Blue, Brown), Blood type (A, B, O), Country.


* **Ordinal Data:** Categories with a **logical order or ranking**, but the mathematical distance between the ranks is not equal or measurable.
* *Examples:* Satisfaction level (Poor, Satisfactory, Excellent), Education level (High School, Bachelors, Masters).



B. Quantitative (Numerical) Data

Data expressed in numbers where mathematical calculations (like averages or subtraction) are valid.

* **Discrete Data:** Countable values that can only take **specific, distinct numbers** (usually whole numbers/integers).
* *Examples:* Number of siblings, pages in a book, cars in a parking lot. (You cannot have 2.5 siblings).


* **Continuous Data:** Measurable values that can take **any real number within a range**, including fractions and decimals.
* *Examples:* Height, weight, temperature, time.



---
3. The Four Levels of Measurement (NOIR)

For numerical data, statisticians use a scale of four levels (from lowest mathematical information to highest):

1. **Nominal:** Categories only. (e.g., Zip codes—they are numbers, but only used as labels).
2. **Ordinal:** Ordered categories. (e.g., Class rank: 1st, 2nd, 3rd).
3. **Interval:** Numerical data with **equal intervals between values, but no "true zero"**. Zero is just an arbitrary marker; negative numbers are possible.
* *Example:* Temperature in Celsius. $0^\circ\text{C}$ does not mean "no temperature at all."


4. **Ratio:** Numerical data with **equal intervals and a "true zero"** (where zero means complete absence of the quantity). Negative values are not possible.
* *Example:* Weight or Income. $0 means literally no money.

```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy as sc

height = np.array ( 
    [70.5, 67.2, 72.1, 68.9, 65.4, 71.3, 69.8, 74.2, 68.1, 70.0,
 66.7, 73.5, 69.1, 68.5, 71.9, 67.8, 70.4, 66.0, 72.8, 69.5,
 65.9, 71.1, 68.3, 70.6, 73.1, 67.4, 69.0, 75.0, 68.7, 70.2]
) 


height

```

```python
rounded_heights = np.round(height).astype(int)

x_coords = []
y_coords = []
counts = {}

for value in rounded_heights:
    if value not in counts:
        counts[value] = 1
    else:
        counts[value] += 1
    
    x_coords.append(value)
    y_coords.append(counts[value]) 
    
plt.figure(figsize=(10, 3))


plt.scatter(x_coords, y_coords, color='black', s=30)

plt.gca().get_yaxis().set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlim(64, 76)
plt.xticks(range(64, 77))
plt.ylim(0, max(y_coords) + 1)
plt.xlabel("Height (Inches)")
plt.tight_layout()
plt.show()
```

Because the raw data is just numbers we count how many times it appears for a range of class interval 

```python
# Define the boundaries of our class intervals
# (We use 75.1 to ensure the maximum value of 75.0 is included in the last interval)


bins = [65.0, 67.0, 69.0, 71.0, 73.0, 75.1]

# Calculate frequencies and boundaries
frequencies, bin_edges = np.histogram(height, bins=bins)


for i in range(len(frequencies)):
    start = bin_edges[i]
    # Represent the intervals as non-overlapping ranges (e.g., 65.0 - 66.9)
    end = bin_edges[i+1] - 0.1 if i < len(frequencies) - 1 else 75.0
    print(f"{start:.1f} - {end:.1f}{'':<6} | {frequencies[i]}")
```

Relative frequency is the fraction or proportion of times a specific value or class interval occurs relative to the total number of observations in a dataset.

While regular frequency tells you the raw count, relative frequency tells you what part of the whole that count represents.m

The Formula$$\text{Relative Frequency} = \frac{f}{n}$$Where:$f = \text{the frequency of a specific value or class interval}$

$n = \text{the total number of observations in the dataset}$

```python
print(f'Total Frequency : {np.sum(frequencies)} ')   #same to number of  obserations
```

```python
relative_freq = [ frequencies[i]/np.sum(frequencies) for i in range(len(frequencies))]
np.round(relative_freq,3)
```

Yes, when you pass a single integer like `bins=3`, **Python chooses the class intervals entirely on its own.** It does this using a simple, standard mathematical formula under the hood.

---


To make the intervals perfectly equal, Python runs three quick steps:

1. **Finds the Extremes:** It looks through your dataset to find the absolute minimum value and the absolute maximum value.
2. **Calculates the Total Range:** It subtracts the minimum from the maximum.
3. **Divides by $k$:** It divides that range by the number of bins ($k$) you requested to find the exact **class width** ($W$).

$$W = \frac{\text{Maximum} - \text{Minimum}}{k}$$

---


In your specific height dataset, Python scans the numbers and finds:

* **Minimum value:** $65.4$
* **Maximum value:** $75.0$

If you type `bins=3`, Python plugs those numbers right into the formula:

$$W = \frac{75.0 - 65.4}{3} = \frac{9.6}{3} = 3.2$$

Because the calculated class width is **3.2**, Python automatically builds these 3 seamless intervals starting exactly at your minimum value:

* **Interval 1:** $65.4 \text{ to } 68.6$
* **Interval 2:** $68.6 \text{ to } 71.8$
* **Interval 3:** $71.8 \text{ to } 75.0$

#### Symmetry


Another important consideration when examining the shape of a distribution is symmetry. A distribution that can be divided into two pieces that are mirror images of one
another is called symmetric.

Example : Bell shaped, uniform distribution

and both of them are unimodal

#### Skewness

A unimodal distribution that is not symmetric is either right skewed or left skewed.
A right-skewed distribution rises to its peak rapidly and comes back toward the horizontal axis more slowly—its “right tail” is longer than its “left tail.” A left-skewed
distribution rises to its peak slowly and comes back toward the horizontal axis more
rapidly—its “left tail” is longer than its “right tail.

A multimodal distribution can be skewed. Skewness is not limited to unimodal curves.

Why Skewness Applies to Multimodal Data

Skewness is simply a measure of asymmetry (how lopsided a graph is). It doesn't care how many hills (peaks) are inside your graph; 

it only cares if the left side and the right side of the entire distribution fail to mirror each other.

####  Median

Another frequently used measure of center is the median. Essentially, the median of a
data set is the number that divides the bottom 50% of the data from the top 50%.


Median of a Data Set
Arrange the data in increasing order.

 If the number of observations is odd, then the median is the observation
exactly in the middle of the ordered list.

 If the number of observations is even, then the median is the mean of the
two middle observations in the ordered list.

In both cases, if we let n denote the number of observations, then the median
is at position (n + 1)/2 in the ordered list

The Mean cares about **MAGNITUDE**: It adds up the actual values of every single data point. It is the center of gravity. If one data point grows larger, the mean must move to balance it.

The Median cares about **RANK (Position)**: It only cares about the order of the points. It is the geographic center—the point that splits your data into a 50/50 split. If the largest value in your dataset becomes 10 times larger, the median doesn't move a single millimeter because the order of the points hasn't changed.

```python
#Estimated mode

idx = np.argmax(counts)

L = edges[idx]                        # Lower limit of modal class (65.0)
f1 = counts[idx]                      # Frequency of modal class (298)
h = edges[idx + 1] - edges[idx]       # Class width (2.0)

f0 = counts[idx - 1] if idx > 0 else 0
f2 = counts[idx + 1] if idx < len(counts) - 1 else 0


estimated_mode = L + ((f1 - f0) / (2 * f1 - f0 - f2)) * h

print(f"The exact estimated Mode is: {estimated_mode:.4f}")
```

**Three-Standard-Deviations Rule**

Almost all the observations in any data set lie within three standard deviations 
to either side of the mean.

$$ x +-   ks $$ where  $$ k = [1,2,3]$$

---

**Chebyshev rule** : For any quantitative data set and any real number $k \geq 1$, at least $1 - \frac{1}{k^2}$ of the observations lie within $k$ standard deviations to either side of the mean, that is, between:
$$\bar{x} - k \cdot s \quad \text{and} \quad \bar{x} + k \cdot s$$ 



This means at least and this occurs for any shape ( whether symmetrical or skewed ) it gives at least assumption 



put $k = 2$ 
$$1 - \frac{1}{2^2} = 75 $$

at least 75 percent data lives in  :
$$\bar{x} - 2 \cdot s \quad \text{and} \quad \bar{x} + 2 \cdot s$$ 

Chebyshev’s rule gives only a minimum for the percentage of
observations that lie within a specified number of standard deviations to either side
of the mean; the actual percentage will usually be higher

---
**Emperical Rule** : 

For any quantitative data set with roughly a bell-shaped distribution, the following properties hold.
Property 1: Approximately 68% of the observations lie within one standard
deviation to either side of the mean, that is, between ¯x − s and ¯x + s.
Property 2: Approximately 95% of the observations lie within two standard
deviations to either side of the mean, that is, between ¯x − 2s and ¯x + 2s.
Property 3: Approximately 99.7% of the observations lie within three standard deviations to either side of the mean, that is, between ¯x − 3s and ¯x + 3s

---

In data analysis, identifying **outliers**—observations that fall well outside the overall pattern of the data—is crucial. An outlier requires special attention 

Common Causes of Outliers
* A measurement or recording error
* An observation from a different population
* An unusual extreme observation 

> **Note:** An extreme observation is not always an outlier; it may indicate **skewness** 

***Example of an Outlier
In a data set of individual wealths (in dollars) of all U.S. residents, the wealth of **Bill Gates** acts as an outlier, representing an unusual extreme observation .***

Handling Outliers
When observing an outlier, try to determine its cause:
1.  **Error-related:** If caused by a measurement/recording error or invalid data point, it can be removed .
2.  **Unexplained:** If no cause is apparent, deciding whether to keep it is a difficult judgment call .


**Quartiles** and the **IQR** (Interquartile Range) are commonly used as diagnostic tools to identify potential outliers, specifically by defining upper and lower limits .



The lower limit and upper limit of a data set are

$$Lower limit = Q1 − 1.5 · IQR$$
$$Upper limit = Q3 + 1.5 · IQR.$$

Observations that lie below the lower limit or above the upper limit are potential
outliers.

Because the population is so large, we use sample mean and sample standard deviation to estimate the population measures.

#### Z-score

According to the formula of **Standard Deviation** it is the average distance from the mean ,

Because we are measuring the distance from the mean, standard deviation is just **euclidean norm** divided by $\sqrt{N}$

$$\mathbf{d} = \mathbf{x} - \boldsymbol{\mu} = \begin{bmatrix} x_1 - \mu \\ x_2 - \mu \\ \vdots \\ x_n - \mu \end{bmatrix}$$

The Euclidean norm (the geometric length) of this deviation vector is:$$\Vert{}\mathbf{d}\Vert{}_2 = \sqrt{\sum_{i=1}^{n} (x_i - \mu)^2}$$

The population standard deviation can be written cleanly using the Euclidean norm of our deviation vector $\mathbf{d}$:$$\sigma = \frac{\Vert{}\mathbf{d}\Vert{}_2}{\sqrt{n}}$$


Now , the **Z-score is given as** : $(x - \mu) / \sigma$.

write out the formula for the vector of all Z-scores, $\mathbf{z}$. To Z-score the dataset, we take the deviation vector and divide it by the scalar value $\sigma$:$$\mathbf{z} = \frac{\mathbf{d}}{\sigma}$$

$$\mathbf{z} = \frac{\mathbf{d}}{\frac{\Vert{}\mathbf{d}\Vert{}_2}{\sqrt{n}}} = \sqrt{n} \cdot \left( \frac{\mathbf{d}}{\Vert{}\mathbf{d}\Vert{}_2} \right)$$

taking any vector and dividing it by its own Euclidean norm creates a unit vector ($\hat{\mathbf{u}}$)—a vector with a length of exactly 1 that points in the same direction.$$\mathbf{z} = \sqrt{n} \cdot \hat{\mathbf{u}}$$

So the  $\sqrt{n}$ is the scalar and the vector $u$ is the unit vector 

Because $u$ is unit vector of $d$. The $u$ vector tells the deviation from the mean but in the range of 1 because its length is 1. Then we multiply by scalar to get the total length of Z score is n = 20 then $Z norm= \sqrt{20} \approx 4.47$

The expression $n\mu$ simply means **$n$ times $\mu$** (the sample size multiplied by the mean). It represents the **total sum of all your data points**.

Think of it this way:
If you have $n = 5$ friends, and their average ($\mu$) wallet balance is $\$10$, how much money do they have total?


$$10 + 10 + 10 + 10 + 10 = 5 \times 10 = \$50$$

Mathematically, the formula for the mean is:


$$\mu = \frac{\sum x_i}{n}$$

If you multiply both sides by $n$ to clear the denominator, you get:


$$\sum x_i = n\mu$$

Also, if you add the constant mean ($\mu$) to itself $n$ times, you get $n\mu$:


$$\sum_{i=1}^{n} \mu = \underbrace{\mu + \mu + \dots + \mu}_{n \text{ times}} = n\mu$$

---

Proof 1: Why the Mean of Z-scores is ALWAYS 0

The mean of any variable is the sum of its values divided by $n$. So the mean of Z ($\mu_z$) is:


$$\mu_z = \frac{1}{n} \sum_{i=1}^{n} z_i$$

Let's look at just the summation part first ($\sum z_i$). Substitute the standard Z-score formula $z_i = \frac{x_i - \mu}{\sigma}$:

$$\sum_{i=1}^{n} z_i = \sum_{i=1}^{n} \left( \frac{x_i - \mu}{\sigma} \right)$$

Since the standard deviation $\sigma$ is a constant denominator for every single term, we can pull it out front:

$$\sum_{i=1}^{n} z_i = \frac{1}{\sigma} \sum_{i=1}^{n} (x_i - \mu)$$

Now, let's expand that remaining summation $\sum(x_i - \mu)$. This means we subtract the mean from every single data point, and add those differences together:

$$\sum_{i=1}^{n} (x_i - \mu) = (x_1 - \mu) + (x_2 - \mu) + \dots + (x_n - \mu)$$

Group all the real data points ($x$'s) together, and group all the $\mu$'s together:

$$\sum_{i=1}^{n} (x_i - \mu) = \underbrace{(x_1 + x_2 + \dots + x_n)}_{\text{Sum of all data points } (\sum x_i)} - \underbrace{(\mu + \mu + \dots + \mu)}_{n \text{ times}}$$

Now substitute what we learned about $n\mu$ above:

* The sum of all $x$ values is **$n\mu$**.
* Adding $\mu$ to itself $n$ times is **$n\mu$**.

$$\sum_{i=1}^{n} (x_i - \mu) = n\mu - n\mu = 0$$

Because the top part equals 0, the whole sum of Z-scores becomes 0:


$$\sum_{i=1}^{n} z_i = \frac{1}{\sigma} \cdot 0 = 0$$

If the sum of all Z-scores is 0, then their mean must be 0:


$$\mu_z = \frac{0}{n} = 0$$

---

Proof 2: Why the Standard Deviation of Z-scores is ALWAYS 1

Since we just proved the mean of $\mathbf{z}$ is $0$, the formula for population variance ($\sigma_z^2$) simplifies significantly. It's just the sum of the squared values divided by $n$:

$$\sigma_z^2 = \frac{1}{n} \sum_{i=1}^{n} z_i^2$$

In linear algebra terms, summing the squares of a vector's components is the exact definition of its **squared Euclidean norm** ($\Vert{}\mathbf{z}\Vert{}_2^2$):

$$\sigma_z^2 = \frac{1}{n} \Vert{}\mathbf{z}\Vert{}_2^2$$

Now, let's bring back your geometric identity: **$\mathbf{z} = \sqrt{n} \cdot \hat{\mathbf{u}}$** (where $\hat{\mathbf{u}}$ is the unit vector).

By absolute definition, a unit vector has a geometric length (norm) of exactly 1:


$$\Vert{}\hat{\mathbf{u}}\Vert{}_2 = 1$$

If we calculate the squared norm of our Z-score vector using your identity, we square both the scalar part and the vector part:

$$\Vert{}\mathbf{z}\Vert{}_2^2 = (\sqrt{n})^2 \cdot \Vert{}\hat{\mathbf{u}}\Vert{}_2^2$$

$$\Vert{}\mathbf{z}\Vert{}_2^2 = n \cdot 1 = n$$

This proves that the total squared length of any Z-score vector is always exactly equal to $n$.

Now, look what happens when we plug this back into our variance equation:

$$\sigma_z^2 = \frac{1}{n} \cdot \left( \Vert{}\mathbf{z}\Vert{}_2^2 \right)$$

$$\sigma_z^2 = \frac{1}{n} \cdot n = 1$$

Since the variance $\sigma_z^2 = 1$, the standard deviation ($\sigma_z$) is just the square root of 1, which is **1**:


$$\sigma_z = \sqrt{1} = 1$$
